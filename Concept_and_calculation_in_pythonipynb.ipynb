#Calculating confidence interval for a sample mean

```python

import numpy as np
from scipy import stats
```

```python
#consider normal distribution
np.random.seed(0)
```

```python
#generate sample from normal distribution
sample_a=stats.norm(0,1).rvs(100)
sample_a[:5]
```

```python
#calculating boundaries
estimate = np.mean(sample_a)
estimate
```

```python
#calculating standard deviation
np.std(sample_a)
```

```python
#calculating standard error
se = np.std(sample_a)/np.sqrt(len(sample_a))
se
```

```python
#calculating sample mean
stats.distributions.norm .ppf(0.025)
```
